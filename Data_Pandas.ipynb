#Data I/O operation , reading and cleaning data using pandas

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Python Projects guide/Pandas/pandas/data/I O operation/stock_data.csv')
# if extra row given skip it  -- pd.read_csv('filename.csv' skiprow=1), or header=1 ---- if header mising then header=None
df
```

```python
df = pd.read_csv('/content/drive/MyDrive/Python Projects guide/Pandas/pandas/data/I O operation/stock_data.csv', na_values=["not available","n.a","n.a."]) #na_values will be ignored
df
```

```python
#data sort now as converted to NaN now use this to convert this
df.fillna(0, inplace=True)
df
```

```python
#error if used in data conversion of Na values in -1 that will result in conversion of whole, we use dictionary to convert that
df = pd.read_csv('/content/drive/MyDrive/Python Projects guide/Pandas/pandas/data/I O operation/stock_data.csv', na_values={
    'eps':["not available","n.a","n.a."],
    'revenue':["not available","n.a","n.a.",-1], #manual entry for columns
    'price': ["not available","n.a","n.a."],
    'tickers': ["not available","n.a","n.a."],
    'people': ["not available","n.a","n.a."]
})
print(df,"\n")
df.fillna(0, inplace=True)
df
```

```python
#write dataframe to to_csv()

df.to_csv('new.csv', index=False) #will be created in root of colab
dd = pd.read_csv('/content/new.csv')
dd
df.to_csv('cutom_colsx.csv', index=False, columns=['eps','price'])
dfx = pd.read_csv('/content/cutom_cols.csv') #created on runtime so data will drop after the session
dfx
```

```python
data_new = pd.read_csv('/content/drive/MyDrive/Python Projects guide/Pandas/pandas/data/I O operation/stock_data.csv', na_values=["not available","n.a","n.a.",-1])
dx = data_new.fillna(method="ffill")
print(data_new, "\n")
dx
```
